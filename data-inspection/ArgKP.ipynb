```python
import pandas as pd
```

```python
DATA_PATH = '../../not-gitted/IBM_Debater_(R)_ArgKP_ACL_2020.v1/ArgKP_dataset.csv'
```

```python
data = pd.read_csv(DATA_PATH)
print(data.info())
data.head()
```

```python
data[['argument']].drop_duplicates().shape
```

```python
truth_data = data[data.label==1]
print(truth_data.info())
truth_data.head()
```

```python
data_groupby_topic = truth_data[['topic', 'argument']].groupby(by=['topic']).count()
data_groupby_topic
```

```python
data_groupby_topic.info()
```

```python
data_groupby_key_point = truth_data[['key_point','argument']].groupby(by=['key_point']).count()
data_groupby_key_point
```

```python
data_groupby_key_point.describe()
```

```python
data_groupby_kp_topic = truth_data[['topic', 'key_point']].groupby(by=['topic']).count()
data_groupby_kp_topic
```

```python
#truth_data[truth_data['key_point'] in truth_data['argument']]
counter = 0
for _,row in truth_data.iterrows():
    if row['key_point'] in row['argument']:
        print(row['key_point'] + ' | '+ row['argument'])
        counter += 1
```

```python
print(f'{counter}\t {counter/len(truth_data)}')
```
