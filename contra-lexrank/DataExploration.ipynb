```python
from DataHandler import DataHandler
import matplotlib.pyplot as plt
```

```python
data = DataHandler()
data.load_json('../../not-gitted/dataset_as_json_file.json')
#data.load_bin('../../not-gitted/dataset_as_json_file.pickle')
len(data.get_arguments())
```

# Topic and query distribution

```python
topics = [a.topic for a in data.get_arguments()]
topics_count = [topics.count(t) for t in topics]
queries = [a.query for a in data.get_arguments()]
queries_count = [queries.count(q) for q in queries]
```

```python
def hist(x, y, title):
    fig, ax = plt.subplots()  

    ax.bar(x, y)

    for tick in ax.get_xticklabels(): 
        tick.set_rotation(75) 

    plt.title(title) 

    plt.figure(figsize=(15,15))
    plt.show()
    #fig.savefig(title+ '.svg')
```

```python
hist(topics, topics_count, 'Topic distribution')
```

```python
hist(queries, queries_count, 'Queries distribution')
```
